```python
import sys
INF = sys.maxsize

def held_karp_tsp(cost_matrix):
    """
    Cài đặt thuật toán Held-Karp (Quy hoạch động) cho bài toán TSP.
    Input: cost_matrix (Ma trận chi phí N x N, cost_matrix[i][j] là chi phí từ i đến j)
    """
    N = len(cost_matrix)

    DP = [[INF] * N for _ in range(1 << N)]
    DP[1][0] = 0
    for mask in range(1, 1 << N):
        for j in range(N):
            if not (mask & (1 << j)):
                continue

            prev_mask = mask ^ (1 << j)

            if prev_mask == 0:
                continue

            for i in range(N):
                if prev_mask & (1 << i):

                    if DP[prev_mask][i] != INF and cost_matrix[i][j] != INF:
                        new_cost = DP[prev_mask][i] + cost_matrix[i][j]
                        DP[mask][j] = min(DP[mask][j], new_cost)


    final_mask = (1 << N) - 1
    min_cost = INF

    for j in range(1, N):
        if DP[final_mask][j] != INF and cost_matrix[j][0] != INF:
            min_cost = min(min_cost, DP[final_mask][j] + cost_matrix[j][0])

    return min_cost

cost_5_cities = [
    [INF, 10, 3, 2, INF], # Từ 0 (A)
    [3, INF, 8, 7, 5],    # Từ 1 (B)
    [2, 4, INF, 6, 9],    # Từ 2 (C)
    [INF, 1, 3, INF, 10], # Từ 3 (D)
    [10, 5, 2, 8, INF]    # Từ 4 (E)
]

result = held_karp_tsp(cost_5_cities)

print(f"Chi phí tối thiểu để đi qua 5 thành phố là: {result}")
```
